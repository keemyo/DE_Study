## 3.Catalyst가 하는일
- Logical Plan을 Physical Plan으로 바꾸는일 
    - Logical Plan
        - 수행해야하는 모든 transformation 단게에 대한 추상화
            - 데이터가 어떻게 변해야 하는지 정의하지만 실제로 어디서 어떻게 동작하는지는 정의하지 않음
        - 컬럼이름, 테이블이름에 대한 검사를 하지 않고 로직을 만듦
            - Unresolved Logical Plan이 만들어짐
    - Pysical Plan
        - Logical Plan이 어떻게 클러스터 위에서 실행될지 정의
            - 실행 전략을 만들고 Cost Plan에 따라 최적화


- Catalyst가 하는일
    1. 분석: DataFrame 객체의 relation을 계산, 칼럼의 타입과 이름 확인
        - 잘못된 테이블 이름, 컬럼을 썼으면 이 단계에서 error
    2. Logical Plan 최적화
        1. 상수로 표현된 표현식을 Compile Time에 게산(x runtime)
        2. Predicate Pushdown: join & filter -> filter & join
            - 테이블이 2개가 조인되고 필터가 되면, Join시 shuffling이 일어나기 때문에, filter를 먼저하는게 효과적
        3. Projection Pruning 연산에 필요한 칼럼만 가져오기
            - 결과값에 필요한 컬럼이 몇개 없다면, 컬럼을 특정해서 가져오는 것
    3. Physical Plan 만들기: Spark에서 실행 가능한 Plan으로 변환
    4. 코드 제너레이션: 최적화된 Pysical Plan을 Java BypteCode로

### Logical Planning 최적화

```sql
SELECT zone_data.Zone, count(*) AS trips
FROM trip_data JOIN zone_data 
ON trip_data.PULocationID = zone_data.LocationID 
WHERE trip_data.hvfhs_license_num = 'HV0003' 
GROUP BY zone_data.Zone order by trips desc;

- trip_data, zone_data를 조인함
- 조건문을 보면
    - 'HV0003'이라는 필터링을 하고
    - 그룹바이를 하고 , 오더링을 함
  
- 로지컬 플랜 예시

## 04. Explain
- spark.sql(query).explain(True)
    - Parsed Logical Plan
    - Analyzed Logical Plan
    - Optimized Logical Plan
    - Physical Plan
- 단계별로, 설명해준다

````python
spark.sql("SELECT zone_data.Zone, count(*) AS trips\
    FROM trip_data JOIN zone_data ON trip_data.PULocationID = zone_data.LocationID \
    WHERE trip_data.hvfhs_license_num = 'HV0003' \
    GROUP BY zone_data.Zone").explain(True)
````
