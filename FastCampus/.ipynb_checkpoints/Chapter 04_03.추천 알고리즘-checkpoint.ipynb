### 01. Alternating Least Squares란?
- Rating Matrix: 빈칸이 많은 Matrix 이건 아래의 두개로 만들어짐
    - User Matrix: 유저
    - Item Matrix: 영화
    
ALS는 둘 중 한 행렬을 고정시키고, 다른 하나를 순차적으로 반복하면서 최적화 해 나가는 것
- 이를 통해 두 행렬의 값을 찾아 나가는거 

1. Item Matrix, User Matrix 랜덤한 값으로 채워짐
2. Item 행렬을 고정, User를 최적화
    - Item, User을 곱했을 때, Rating에 최대한 비슷해지게 맞춰줌
3. User 행렬을 고정, Item을 최적화
    - 여러번 반복했을 때, 두 값을 합쳤을 때 Rating Matrix와 최대한 비슷해짐
4. 이 때, Rating Matrix에서 비어 있는 값이 차 있어서 유추 가능
